# <span style="color:#0f766e; font-weight:700;">1. Czym jest DataFrame?</span>

`DataFrame` to podstawowa struktura danych w `pandas`.

Najprościej:
- jest to **tabela** z wierszami i kolumnami,
- każda kolumna ma nazwę,
- każda kolumna ma typ danych (`dtype`),
- można na niej wykonywać filtrowanie, agregacje, łączenie tabel i przekształcenia.

W praktyce `DataFrame` łączy w sobie cechy:
- arkusza kalkulacyjnego,
- tabeli SQL,
- obiektu programistycznego, na którym można budować cały pipeline analityczny.

## <span style="color:#0f766e; font-weight:700;">Jak myśleć o DataFrame?</span>

- **wiersz** = jedna obserwacja / rekord,
- **kolumna** = jedna zmienna / cecha,
- **indeks** = etykieta wiersza,
- **dtype** = techniczna reprezentacja danych w pamięci.

Przykłady:
- tabela pracowników,
- tabela transakcji,
- wyniki ankiety,
- logi eksperymentu.


# <span style="color:#0f766e; font-weight:700;">2. Najważniejsze operacje na DataFrame</span>

## <span style="color:#0f766e; font-weight:700;">Tworzenie i wczytywanie</span>

Najczęściej używane wejścia:
- `pd.read_csv()` - wczytuje dane z pliku CSV do `DataFrame`.
- `pd.read_excel()` - wczytuje dane z pliku Excel (`.xlsx`, `.xls`).
- `pd.read_json()` - wczytuje dane zapisane w formacie JSON.
- `pd.DataFrame()` - tworzy `DataFrame` ręcznie, np. ze słownika lub listy.

Szybka diagnostyka po wczytaniu:
- `head()` - pokazuje pierwsze wiersze tabeli.
- `tail()` - pokazuje ostatnie wiersze tabeli.
- `sample()` - losuje przykładowe rekordy do szybkiego podglądu.
- `shape` - zwraca liczbę wierszy i kolumn.
- `info()` - pokazuje typy kolumn, braki danych i podstawową strukturę.
- `dtypes` - zwraca typ danych każdej kolumny.
- `describe()` - liczy podstawowe statystyki opisowe dla kolumn liczbowych.

Przykład:
```python
import pandas as pd

df = pd.read_csv('plik.csv')
df.head()
df.info()
```

## <span style="color:#0f766e; font-weight:700;">Wybieranie danych</span>

Najważniejsze narzędzia:
- wybór kolumn: `df[['kol1', 'kol2']]` - zwraca nowy `DataFrame` z wybranymi kolumnami.
- wybór jednej kolumny: `df['kol1']` - zwraca pojedynczą serię (`Series`).
- wybór etykietami: `df.loc[...]` - wybiera dane po nazwach wierszy i kolumn.
- wybór po pozycjach: `df.iloc[...]` - wybiera dane po numerach pozycji.
- filtrowanie logiczne: `df[df['wiek'] > 30]` - zostawia tylko rekordy spełniające warunek.
- zapytania tekstowe: `df.query(...)` - pozwala filtrować dane składnią podobną do SQL.

Przykład:
```python
df.loc[df['salary_pln'] > 10000, ['department', 'salary_pln']]
```

## <span style="color:#0f766e; font-weight:700;">Czyszczenie i transformacje</span>

Najważniejsze metody:
- `rename()` - zmienia nazwy kolumn lub indeksu.
- `drop()` - usuwa wskazane kolumny albo wiersze.
- `dropna()` - usuwa wiersze lub kolumny z brakami danych.
- `fillna()` - uzupełnia braki danych wskazaną wartością.
- `astype()` - konwertuje kolumnę na inny typ danych.
- `assign()` - dodaje nową kolumnę lub kilka kolumn w jednym kroku.
- `replace()` - zamienia wskazane wartości na inne.
- `sort_values()` - sortuje dane według wartości w kolumnie.
- `sort_index()` - sortuje dane według indeksu.
- `reset_index()` - przywraca zwykły indeks liczbowy.
- `set_index()` - ustawia wybraną kolumnę jako indeks.

Przykład:
```python
df = df.rename(columns={'salary_pln': 'salary'})
df = df.sort_values('salary', ascending=False)
```

## <span style="color:#0f766e; font-weight:700;">Agregacje i grupowanie</span>

Najważniejsze:
- `value_counts()` - liczy częstość występowania wartości w kolumnie.
- `groupby()` - dzieli dane na grupy według jednej lub kilku kolumn.
- `agg()` - pozwala policzyć kilka agregacji naraz.
- `mean()` - liczy średnią.
- `median()` - liczy medianę.
- `sum()` - liczy sumę.
- `min()` - zwraca wartość minimalną.
- `max()` - zwraca wartość maksymalną.
- `std()` - liczy odchylenie standardowe.
- `nunique()` - liczy liczbę unikalnych wartości.

Przykład:
```python
df.groupby('department').agg(
    salary_mean=('salary_pln', 'mean'),
    people=('employee_id', 'count')
)
```

## <span style="color:#0f766e; font-weight:700;">Łączenie tabel</span>

Najważniejsze:
- `merge()` - łączy dwie tabele po wspólnym kluczu.
- `join()` - wygodne łączenie po indeksie lub kluczach.
- `concat()` - skleja tabele pionowo albo poziomo.

Przykład:
```python
df = employees.merge(departments, on='department', how='left')
```

## <span style="color:#0f766e; font-weight:700;">Zmiana kształtu tabeli</span>

Najważniejsze:
- `pivot_table()` - tworzy tabelę przestawną z agregacją.
- `pivot()` - przekształca układ danych bez agregowania.
- `melt()` - zamienia układ szeroki na długi.

Przykład:
```python
scores_long = scores.melt(
    id_vars='employee_id',
    var_name='quarter',
    value_name='score'
)
```

## <span style="color:#0f766e; font-weight:700;">Operacje tekstowe i daty</span>

Bardzo ważne accesory:
- `.str` - udostępnia metody do pracy na tekście w kolumnach.
- `.dt` - udostępnia metody do pracy na datach i czasie.

Najczęściej używane:
- `str.lower()` - zamienia tekst na małe litery.
- `str.strip()` - usuwa zbędne spacje z początku i końca tekstu.
- `str.contains()` - sprawdza, czy tekst zawiera wzorzec.
- `to_datetime()` - konwertuje tekst na datę.
- `dt.year` - wyciąga rok z daty.
- `dt.month` - wyciąga numer miesiąca.
- `dt.day_name()` - zwraca nazwę dnia tygodnia.

Przykład:
```python
df['hire_date'] = pd.to_datetime(df['hire_date'])
df['year'] = df['hire_date'].dt.year
```

## <span style="color:#0f766e; font-weight:700;">Wykonywalne przykłady</span>

Poniżej znajdziesz wykonywalne przykłady kodu dla każdej najważniejszej grupy operacji na `DataFrame`. Każdą komórkę można uruchomić niezależnie.


## <span style="color:#0f766e; font-weight:700;">Przykład 1: Wczytywanie i szybka diagnostyka</span>

Ta komórka pokazuje, jak wczytać dane i natychmiast sprawdzić ich strukturę: `head`, `tail`, `sample`, `shape`, `info`, `dtypes`, `describe`.


````python
import pandas as pd

# Wczytujemy plik CSV do obiektu DataFrame.
# Parametr: ścieżka tekstowa 'dane_lekcja3/employees.csv' wskazuje plik źródłowy.
df = pd.read_csv('dane_lekcja3/employees.csv')

# head() pokazuje pierwsze 5 wierszy.
# Brak parametru oznacza domyślnie n=5.
print('HEAD:')
print(df.head())
print()

# tail() pokazuje ostatnie 5 wierszy.
# Brak parametru oznacza domyślnie n=5.
print('TAIL:')
print(df.tail())
print()

# sample() losuje przykładowe rekordy.
# Parametr 3 oznacza liczbę losowanych wierszy.
# Parametr random_state=42 ustawia ziarno losowości, więc wynik będzie powtarzalny.
print('SAMPLE (3):')
print(df.sample(3, random_state=42))
print()

# shape nie jest funkcją, tylko atrybutem.
# Zwraca krotkę: (liczba_wierszy, liczba_kolumn).
print('SHAPE:', df.shape)
print()

# info() pokazuje strukturę tabeli.
# Widzimy liczbę niepustych wartości, typy danych i przybliżone zużycie pamięci.
print('INFO:')
df.info()
print()

# dtypes zwraca serię z typami każdej kolumny.
print('DTYPES:')
print(df.dtypes)
print()

# describe() liczy statystyki opisowe.
# Parametr include='all' każe uwzględnić także kolumny tekstowe.
# .T transponuje wynik, żeby łatwiej czytać statystyki dla każdej kolumny w wierszu.
# head(10) ogranicza wydruk do pierwszych 10 wierszy wyniku.
print('DESCRIBE:')
print(df.describe(include='all').T.head(10))
````

## <span style="color:#0f766e; font-weight:700;">Przykład 2: Wybieranie kolumn, filtrowanie, `loc`, `iloc`, `query`</span>

Ta komórka pokazuje różnicę między wyborem kolumn, filtrowaniem warunkiem, wyborem po etykietach (`loc`) i po pozycjach (`iloc`).


````python
import pandas as pd

# Wczytujemy dane pracowników do DataFrame.
df = pd.read_csv('dane_lekcja3/employees.csv')

# Wybór kilku kolumn przez listę nazw.
# Podwójne nawiasy kwadratowe zwracają DataFrame, a nie pojedynczą serię.
subset = df[['employee_id', 'department', 'city', 'salary_pln']]
print('WYBRANE KOLUMNY:')
print(subset.head())
print()

# Filtrowanie logiczne.
# Warunek df['salary_pln'] > 12000 tworzy serię wartości True/False.
# DataFrame[warunek] zostawia tylko wiersze, dla których warunek jest True.
high_salary = df[df['salary_pln'] > 12000]
print('PENSJA > 12000:')
print(high_salary[['employee_id', 'department', 'salary_pln']].head())
print()

# loc wybiera dane po etykietach.
# Pierwszy argument 0:4 oznacza zakres wierszy po indeksie.
# Drugi argument to lista nazw kolumn, które chcemy pokazać.
print('LOC (wiersze 0:4, konkretne kolumny):')
print(df.loc[0:4, ['first_name', 'last_name', 'salary_pln']])
print()

# iloc wybiera dane po pozycjach liczbowych.
# Pierwszy argument 0:5 oznacza wiersze o pozycjach 0-4.
# Drugi argument 1:4 oznacza kolumny o pozycjach 1-3.
print('ILOC (wiersze 0:5, kolumny 1:4):')
print(df.iloc[0:5, 1:4])
print()

# query() pozwala zapisać filtr jako czytelny tekst.
# Wewnątrz stringa odwołujemy się do nazw kolumn bez df['...'].
# Wynik dodatkowo ograniczamy do kilku kolumn.
print('QUERY department == analytics and salary_pln > 10000:')
print(
    df.query("department == 'analytics' and salary_pln > 10000")
      [['employee_id', 'department', 'salary_pln']]
      .head()
)
````

## <span style="color:#0f766e; font-weight:700;">Przykład 3: Transformacje, zmiana nazw, nowe kolumny, sortowanie, indeks</span>

Ta komórka pokazuje typowy pipeline przekształceń na `DataFrame`.


````python
import pandas as pd

# Wczytujemy dane pracowników.
df = pd.read_csv('dane_lekcja3/employees.csv')

# rename() zmienia nazwy kolumn.
# Parametr columns przyjmuje słownik: stara_nazwa -> nowa_nazwa.
df = df.rename(columns={'salary_pln': 'salary'})

# assign() dodaje nowe kolumny w jednym kroku.
# annual_salary to pensja miesięczna * 12.
# bonus_value to liczbowy odpowiednik procentowego bonusu.
df = df.assign(
    annual_salary=df['salary'] * 12,
    bonus_value=df['salary'] * df['bonus_pct']
)

# replace() podmienia wskazane wartości.
# Tutaj zamieniamy skrót 'UoP' na pełniejszą nazwę.
# Parametr przekazujemy jako słownik: stara_wartość -> nowa_wartość.
df['contract_type'] = df['contract_type'].replace({'UoP': 'umowa_o_prace'})

# sort_values() sortuje tabelę po wartości w kolumnie.
# Parametr 'annual_salary' wskazuje kolumnę sortowania.
# Parametr ascending=False ustawia sortowanie malejące.
sorted_df = df.sort_values('annual_salary', ascending=False)
print('PO SORTOWANIU:')
print(sorted_df[['employee_id', 'salary', 'annual_salary']].head())
print()

# set_index() ustawia wskazaną kolumnę jako indeks wierszy.
# Parametr 'employee_id' mówi, która kolumna ma stać się indeksem.
indexed = sorted_df.set_index('employee_id')
print('PO SET_INDEX:')
print(indexed.head(3))
print()

# reset_index() przywraca zwykły indeks liczbowy i zwraca dawny indeks jako kolumnę.
reset_back = indexed.reset_index()
print('PO RESET_INDEX:')
print(reset_back.head(3))
````

## <span style="color:#0f766e; font-weight:700;">Przykład 4: `value_counts`, `groupby`, `agg`, `nunique`</span>

Ta komórka pokazuje najczęstsze operacje podsumowujące na `DataFrame`.


````python
import pandas as pd

# Wczytujemy tabelę pracowników.
df = pd.read_csv('dane_lekcja3/employees.csv')

# value_counts() liczy wystąpienia każdej wartości w kolumnie.
print('LICZEBNOSC DEPARTAMENTOW:')
print(df['department'].value_counts())
print()

# nunique() liczy liczbę unikalnych wartości w kolumnie.
print('LICZBA UNIKALNYCH MIAST:', df['city'].nunique())
print()

# groupby('department') dzieli dane na grupy według działu.
# agg() liczy wiele agregacji jednocześnie.
# Każdy wpis ma postać: nowa_nazwa_kolumny=(kolumna_wejściowa, funkcja_agregująca).
summary = df.groupby('department').agg(
    employees=('employee_id', 'count'),
    salary_mean=('salary_pln', 'mean'),
    salary_median=('salary_pln', 'median'),
    performance_mean=('performance_score', 'mean')
)

# sort_values() porządkuje wynik po średniej pensji malejąco.
summary = summary.sort_values('salary_mean', ascending=False)

print('PODSUMOWANIE WG DEPARTMENT:')
print(summary)
````

## <span style="color:#0f766e; font-weight:700;">Przykład 5: `merge`, `join`, `concat`</span>

Ta komórka pokazuje trzy różne sposoby łączenia danych.


````python
import pandas as pd

# Wczytujemy dwie tabele, które będziemy łączyć.
employees = pd.read_csv('dane_lekcja3/employees.csv')
departments = pd.read_csv('dane_lekcja3/departments.csv')

# merge() łączy tabele po wspólnym kluczu.
# Parametr on='department' mówi, że kluczem jest kolumna department po obu stronach.
# Parametr how='left' oznacza: zachowaj wszystkie rekordy z lewej tabeli (employees).
# Parametr indicator=True dodaje kolumnę _merge z informacją o dopasowaniu rekordów.
merged = employees.merge(departments, on='department', how='left', indicator=True)
print('MERGE:')
print(merged[['employee_id', 'department', 'director', '_merge']].head())
print()

# join() jest wygodny, gdy łączymy po indeksie.
# Najpierw ustawiamy w obu tabelach department jako indeks.
emp_idx = employees.set_index('department')
dep_idx = departments.set_index('department')

# Parametr how='left' działa analogicznie jak w merge.
# Parametr rsuffix='_dept' dodaje sufiks do kolumn konfliktujących nazwami po prawej stronie.
joined = emp_idx.join(dep_idx, how='left', rsuffix='_dept')
print('JOIN:')
print(joined[['employee_id', 'director', 'budget_k_pln']].head())
print()

# concat() skleja obiekty razem.
# Tutaj bierzemy dwa fragmenty tej samej tabeli.
part_a = employees.head(2)
part_b = employees.tail(2)

# Parametr axis=0 oznacza sklejanie pionowe, czyli dokładanie wierszy.
concatenated = pd.concat([part_a, part_b], axis=0)
print('CONCAT:')
print(concatenated[['employee_id', 'department']])
````

## <span style="color:#0f766e; font-weight:700;">Przykład 6: `melt`, `pivot`, `pivot_table`</span>

Ta komórka pokazuje zmianę kształtu danych: z układu szerokiego na długi i budowę tabeli przestawnej.


````python
import pandas as pd

# Wczytujemy dane w układzie szerokim (wide).
scores = pd.read_csv('dane_lekcja3/employee_scores_wide.csv')

# melt() zamienia wiele kolumn wyników na układ długi.
# Parametr id_vars='employee_id' wskazuje kolumnę identyfikującą rekord.
# Parametr var_name='quarter' tworzy nową kolumnę z dawnymi nazwami kolumn.
# Parametr value_name='score' tworzy kolumnę z wartościami.
scores_long = scores.melt(
    id_vars='employee_id',
    var_name='quarter',
    value_name='score'
)
print('MELT (wide -> long):')
print(scores_long.head())
print()

# Przygotowujemy małą, ale kompletną próbkę do demonstracji pivot().
# Bierzemy 3 identyfikatory, dla których istnieją wszystkie 4 kwartały.
example_ids = scores['employee_id'].head(3).tolist()
small = scores_long[scores_long['employee_id'].isin(example_ids)].copy()

# pivot() zamienia układ długi z powrotem na szeroki.
# Parametr index='employee_id' ustawia wiersze.
# Parametr columns='quarter' tworzy kolumny z nazw kwartałów.
# Parametr values='score' wskazuje, jakie wartości wypełniają komórki.
pivoted = small.pivot(index='employee_id', columns='quarter', values='score')
print('PIVOT (long -> wide, kompletna probka):')
print(pivoted)
print()

# Wczytujemy drugą tabelę do demonstracji pivot_table().
hours = pd.read_csv('dane_lekcja3/monthly_hours.csv')

# pivot_table() działa jak tabela przestawna z agregacją.
# Parametr index='month' ustawia wiersze tabeli wynikowej.
# Parametr values='billable_hours' wskazuje kolumnę do agregacji.
# Parametr aggfunc='mean' mówi, że liczymy średnią.
pivot_tbl = hours.pivot_table(
    index='month',
    values='billable_hours',
    aggfunc='mean'
)
print('PIVOT_TABLE (srednie billable_hours wg month):')
print(pivot_tbl)
````

## <span style="color:#0f766e; font-weight:700;">Przykład 7: `.str`, `pd.to_datetime()`, `.dt`</span>

Ta komórka pokazuje pracę z tekstem i datami na kolumnach `DataFrame`.


````python
import pandas as pd

# Wczytujemy dane pracowników.
df = pd.read_csv('dane_lekcja3/employees.csv')

# Operacje tekstowe przez accessor .str.
# str.lower() zamienia tekst na małe litery.
# str.strip() usuwa zbędne spacje z początku i końca tekstu.
df['first_name_clean'] = df['first_name'].str.lower().str.strip()
df['last_name_clean'] = df['last_name'].str.lower().str.strip()

# Łączymy dwie kolumny tekstowe w jedną nową kolumnę full_name.
df['full_name'] = df['first_name_clean'] + ' ' + df['last_name_clean']

# str.contains('a') sprawdza, czy w tekście występuje litera 'a'.
# Wynik to seria wartości True/False.
df['is_name_with_a'] = df['full_name'].str.contains('a')

print('OPERACJE TEKSTOWE:')
print(df[['first_name', 'last_name', 'full_name', 'is_name_with_a']].head())
print()

# pd.to_datetime() konwertuje tekst na typ daty.
# Parametr przekazujemy jako serię, którą chcemy przekonwertować.
df['hire_date'] = pd.to_datetime(df['hire_date'])

# Accessor .dt udostępnia operacje na dacie.
# dt.year wyciąga rok.
# dt.month wyciąga numer miesiąca.
# dt.day_name() zwraca nazwę dnia tygodnia.
df['hire_year'] = df['hire_date'].dt.year
df['hire_month'] = df['hire_date'].dt.month
df['hire_day_name'] = df['hire_date'].dt.day_name()

print('OPERACJE NA DATACH:')
print(df[['hire_date', 'hire_year', 'hire_month', 'hire_day_name']].head())
````

## <span style="color:#0f766e; font-weight:700;">Zadanie 4</span>

Przekształć tabelę: zmień nazwę kolumny `salary_pln` na `salary`, dodaj kolumnę `annual_salary`, a następnie posortuj dane malejąco po `annual_salary`.

- `rename()` - zmienia nazwę kolumny.
- `assign()` albo przypisanie kolumny - dodaje nową kolumnę do tabeli.
- `sort_values()` - sortuje rekordy według wartości w kolumnie.

## <span style="color:#0f766e; font-weight:700;">Zadanie 9</span>

Na podstawie `monthly_hours.csv` zbuduj tabelę przestawną pokazującą średnie `billable_hours` w podziale na `month` i `employee_id` albo `department` (po wcześniejszym merge).

Wymagane funkcje/parametry:
- `pd.read_csv()` - wczytuje dane z pliku CSV.
- `merge()` - łączy tabelę godzin z inną tabelą, np. pracowników.
- `pivot_table()` - buduje tabelę przestawną z agregacją.
- parametr `index` - ustawia wiersze tabeli przestawnej.
- parametr `columns` - ustawia kolumny tabeli przestawnej.
- parametr `values` - wskazuje, którą zmienną agregować.
- parametr `aggfunc` - określa, jaką funkcję agregującą zastosować.

Kryterium zaliczenia:
- zadanie wykonane zgodnie z poleceniem,
- wynik pokazany w notebooku.

Checkpoint:
- pokaż prowadzącemu wynik końcowy tej sekcji.

Dokumentacja:
- https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html


````python
# tutaj dodaj swój kod
monthly_hours = pd.read_csv('./dane_lekcja3/monthly_hours.csv')
monthly_hours = monthly_hours.merge(employees, on='employee_id', how='left', indicator=False)
monthly_hours_pivot = monthly_hours.pivot_table(index=['month', 'employee_id'], columns=['department'], values='billable_hours', aggfunc='mean')
monthly_hours_pivot
````
