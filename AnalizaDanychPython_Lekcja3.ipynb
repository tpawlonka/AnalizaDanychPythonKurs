## <span style="color:#0f766e; font-weight:700;">Zadanie 1</span>

Wczytaj wszystkie pliki projektu i sprawdź podstawowe informacje o tabelach (wymiary, typy, braki).

Wymagane biblioteki:
- `pandas`
- `json`

Wymagane funkcje/parametry:
- `pd.read_csv()`
- `json.load()` / `json.loads()`
- `DataFrame.shape`
- `DataFrame.info()`
- `DataFrame.isna()`
- `DataFrame.sum()`


```python
# tutaj dodaj swój kod
import pandas as pd
import json as js

city_reference_dirty = pd.read_csv('./dane_lekcja3/city_reference.csv')
participants_dirty = pd.read_csv('./dane_lekcja3/participants_dirty.csv', na_values=['brak', '.'])
with open('./dane_lekcja3/project_questions.json', 'r') as json_raw:
    json = js.load(json_raw)
    project_questions = pd.DataFrame(json.values())
trials_dirty = pd.read_csv('./dane_lekcja3/trials_dirty.csv', na_values='brak')

dataframes_dirty = [city_reference_dirty, participants_dirty, project_questions, trials_dirty]
for df in dataframes_dirty:
    print(df.shape)
    print(df.info())
    print(df.isna().sum())
    print('\n')
```

## <span style="color:#0f766e; font-weight:700;">Zadanie 2</span>

Znormalizuj nazwy kategorii w kolumnach `group`, `gender`, `education`, `condition`, `difficulty` (wielkość liter, warianty tekstowe).

Wymagane biblioteki:
- `pandas`

Wymagane funkcje/parametry:
- `Series.str.lower()`
- `Series.str.strip()`
- `Series.replace()`
- `Series.map()`
- `Series.value_counts()`


```python
# tutaj dodaj swój kod]
print(f'{participants_dirty['gender'].value_counts()}\n')
print(f'{participants_dirty['group'].value_counts()}\n')
print(f'{participants_dirty['education'].value_counts()}\n')
print(f'{trials_dirty['condition'].value_counts()}\n')
print(f'{trials_dirty['difficulty'].value_counts()}\n')

participants = participants_dirty.copy()
trials = trials_dirty.copy()
participants['gender'] = participants_dirty['gender'].str.lower().str.strip().replace(to_replace=['f', 'k', 'kobieta'], value='female').replace(to_replace=['m', 'mezczyzna'], value='male')
participants['group'] = participants_dirty['group'].str.lower().str.strip()
participants['education'] = participants_dirty['education'].str.lower().str.strip().map(lambda x: 'licencjat' if x in ['lic'] else 'magisterium' if x in ['mgr', 'magister'] else x)
trials['condition'] = trials_dirty['condition'].str.lower().str.strip()
trials['difficulty'] = trials_dirty['difficulty'].str.lower().str.strip()
city_reference = city_reference_dirty.copy()

print('\n')
print(f'{participants['gender'].value_counts()}\n')
print(f'{participants['group'].value_counts()}\n')
print(f'{participants['education'].value_counts()}\n')
print(f'{trials['difficulty'].value_counts()}\n')
```
